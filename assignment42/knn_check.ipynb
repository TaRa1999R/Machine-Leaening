We use apple & banana assignment to check the K.N.N algorithm

```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from knn import KNN
```

Define a function to produce dataset

- apple width ≈ 6 cm
- apple length ≈ 6 cm

- banana width ≈ 4 cm
- banana length ≈ 8 cm

```python
def generate_dataset (n) :
    apple_width = np.random.normal (6 , 1 , n)
    apple_length = np.random.normal (6 , 1 , n)

    banana_width = np.random.normal (4 , 1 , n)
    banana_length = np.random.normal (8 , 1 , n)

    width = np.concatenate ((apple_width , banana_width))
    length = np.concatenate ((apple_length , banana_length))

    x = np.array ([width , length]).T

    y_apple = np.zeros (n , dtype = 'int')
    y_banana = np.ones (n , dtype = 'int')
    y = np.concatenate ((y_apple , y_banana))

    return x , y

```

Generate x-train and y_train width defined function

```python
X_train , Y_train = generate_dataset (500)
```

- apple => red
- banana => yellow

```python
cmap = ['red' , 'gold']
plt.scatter (X_train[: , 0] , X_train[: , 1] , c = Y_train , cmap = ListedColormap (cmap) , marker = '.')

plt.title ('Apple & Banana')
plt.xlabel ('width')
plt.ylabel ('length')

plt.savefig ('outputs/apple, banana and watermelon.png')
plt.show ()
```

Use KNN

```python
knn = KNN (k=5)
knn.fit (X_train , Y_train)
```

new data

```python
new_1 = np.array ([3 , 8.4])
new_2 = np.array ([6 , 6])
new = [new_1 , new_2]

outputs = knn.predict (new)

for output in outputs :
    if output == 0 :
        print ('Apple')

    else :
        print ('Banana')
```

check accuracy

```python
X_test , Y_test = generate_dataset (50)

knn.evaluate (X_test , Y_test)
```

compare with sklearn

```python
from sklearn.neighbors import KNeighborsClassifier

knn2 = KNeighborsClassifier (n_neighbors=5)
knn2.fit (X_train , Y_train)
print (knn2.predict (new))
print (knn2.score (X_test , Y_test))
```

So our K.N.N module is working perfectly
